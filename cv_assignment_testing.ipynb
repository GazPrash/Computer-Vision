```python
ClockwiseRotMatrix = lambda x : numpy.array([math.cos(x), -math.sin(x), math.sin(x), math.cos(x)]).reshape(2, 2)
antiClockwiseRotMatrix = lambda x : numpy.array([math.cos(x), math.sin(x), -math.sin(x), math.cos(x)]).reshape(2, 2)
# creating two lambda functions that return a 2x2 matrix that will be used for rotation for any given input angle x, for clockwise and anticlockwise respectively
```

```python
angleRot = 80*math.pi/180 # choose a random angle b/w 0 & 3pi/2 rad 
MAX_LEN = int(math.sqrt(height*height + width*width))

rotated_img = numpy.zeros((height, width, 3))
centreX, centreY = width//2, height//2


for row_ind, row in enumerate(img):
    for col_ind, col in enumerate(row):
        new_indices = numpy.matmul(antiClockwiseRotMatrix(angleRot), numpy.array([row_ind - centreX, col_ind - centreY]).reshape(2, 1))
        new_wid = round((float(new_indices[0])) + centreX)
        new_hi = round((float(new_indices[1])) + centreY)

        if (new_wid >= 0 and new_hi >= 0) and (new_wid < width and new_hi < height):
            # print(row_ind, col_ind, new_wid, new_hi)
            rotated_img[row_ind, col_ind][:] = img[new_hi, new_wid][:]


cv2.imwrite("images/rotated.png", rotated_img)
print(f"Rotated image by : {math.degrees(angleRot)} degrees")
Image("images/rotated.png")
```
